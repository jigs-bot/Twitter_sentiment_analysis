```python
def avg_w2v(doc):
    return np.mean([t[word] for word in doc if word in t.index_to_key],axis=0)
```

```python
from tqdm import tqdm
```

```python
X=[]
for i in tqdm(range(len(w))):
    X.append(avg_w2v(w[i]))
```
